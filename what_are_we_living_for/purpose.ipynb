```python
# Source: https://stackoverflow.com/questions/16645799/how-to-create-a-word-cloud-from-a-corpus-in-python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text = (12 * " Love") + (8 * " Sex") + (10 * " Happiness") + (7 * " Power") + (10 * " Peace") + (3 * " Others")
wordcloud = WordCloud(font_path='/Library/Fonts/Phosphate.ttc', background_color='white', relative_scaling = 0.3).generate(text)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

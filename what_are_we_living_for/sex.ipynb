```python
import matplotlib.pyplot as plt

Sex = 'Male', 'Female', 'Others'
sizes = [22, 20, 8]
Colors = ['yellowgreen', 'lightcoral', 'lightblue']
explode = (0, 0, 0.1)

## :autopct: https://stackoverflow.com/questions/6170246/how-do-i-use-matplotlib-autopct 

plt.pie(sizes, explode=explode, labels=Sex, shadow=True, autopct='%1.1f%%', colors=Colors, startangle=90)
plt.axis('equal')
plt.show()
```
